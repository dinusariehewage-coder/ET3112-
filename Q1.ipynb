```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
import os


image_path = '/Assignment 01/runway.png'



f = cv.imread(image_path, cv.IMREAD_GRAYSCALE)


if f is None:
    raise FileNotFoundError("Image not found. Check the image path and filename.")

# Normalize image to range [0, 1]
f = f / 255.0


# (a) Gamma correction with γ = 0.5

g1 = f ** 0.5


# (b) Gamma correction with γ = 2

g2 = f ** 2

# (c) Contrast Stretching (piecewise linear)
# r1 = 0.2, r2 = 0.8

r1, r2 = 0.2, 0.8
g3 = np.where(f < r1, 0,
              np.where(f <= r2, (f - r1) / (r2 - r1), 1))



plt.figure(figsize=(10, 8))

plt.subplot(2, 2, 1)
plt.imshow(f, cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(2, 2, 2)
plt.imshow(g1, cmap='gray')
plt.title('Gamma Correction (γ = 0.5)')
plt.axis('off')

plt.subplot(2, 2, 3)
plt.imshow(g2, cmap='gray')
plt.title('Gamma Correction (γ = 2)')
plt.axis('off')

plt.subplot(2, 2, 4)
plt.imshow(g3, cmap='gray')
plt.title('Contrast Stretching')
plt.axis('off')

plt.tight_layout()
plt.show()
```
