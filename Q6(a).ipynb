```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D


sigma = 2
size = 51   # kernel size (odd number)

# Coordinate grid
ax = np.arange(-(size//2), size//2 + 1)
x, y = np.meshgrid(ax, ax)


# Gaussian function G(x, y)

G = (1 / (2 * np.pi * sigma**2)) * np.exp(-(x**2 + y**2) / (2 * sigma**2))

# First-order derivatives

Gx = -(x / sigma**2) * G   # ∂G/∂x
Gy = -(y / sigma**2) * G   # ∂G/∂y


# Plot Gaussian

fig = plt.figure(figsize=(18, 5))

ax1 = fig.add_subplot(131, projection='3d')
ax1.plot_surface(x, y, G, cmap='viridis')
ax1.set_title('Gaussian G(x, y)')
ax1.set_xlabel('x')
ax1.set_ylabel('y')

# -----------------------------
# Plot ∂G/∂x
# -----------------------------
ax2 = fig.add_subplot(132, projection='3d')
ax2.plot_surface(x, y, Gx, cmap='coolwarm')
ax2.set_title('Derivative ∂G/∂x')
ax2.set_xlabel('x')
ax2.set_ylabel('y')

# -----------------------------
# Plot ∂G/∂y
# -----------------------------
ax3 = fig.add_subplot(133, projection='3d')
ax3.plot_surface(x, y, Gy, cmap='coolwarm')
ax3.set_title('Derivative ∂G/∂y')
ax3.set_xlabel('x')
ax3.set_ylabel('y')

plt.tight_layout()
plt.show()
```
