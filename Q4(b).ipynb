```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

# Copy grayscale image
gray_equalized = gray.copy()

# Get indices of foreground pixels
foreground_indices = np.where(binary_mask > 0)

# Extract foreground pixels (1D array)
foreground_pixels = gray[foreground_indices]

# OpenCV equalizeHist works on 2D images, so reshape temporarily
foreground_pixels_2d = foreground_pixels.reshape(-1, 1)

# Apply histogram equalization
equalized_pixels = cv.equalizeHist(foreground_pixels_2d)

# Flatten back to 1D to assign
equalized_pixels = equalized_pixels.flatten()

# Assign equalized pixels back to the foreground region
gray_equalized[foreground_indices] = equalized_pixels

# Display original and equalized images
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(gray, cmap='gray')
plt.title('Original Grayscale')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(gray_equalized, cmap='gray')
plt.title('Histogram Equalized (Foreground)')
plt.axis('off')
plt.show()
```
