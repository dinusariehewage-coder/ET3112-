```python
import numpy as np

# Kernel parameters
kernel_size = 5
sigma = 2

# Create a grid of (x,y) coordinates
ax = np.arange(-(kernel_size // 2), kernel_size // 2 + 1, 1)
x, y = np.meshgrid(ax, ax)

# Compute Gaussian function
gaussian_kernel = np.exp(-(x**2 + y**2) / (2 * sigma**2))

# Normalize so sum = 1
gaussian_kernel /= np.sum(gaussian_kernel)

print("Normalized 5x5 Gaussian Kernel (σ=2):\n", gaussian_kernel)
```
