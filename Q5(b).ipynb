```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Kernel parameters
kernel_size_3d = 51
sigma_3d = 2

# Grid
ax = np.arange(-(kernel_size_3d // 2), kernel_size_3d // 2 + 1, 1)
x, y = np.meshgrid(ax, ax)

# Compute Gaussian
gaussian_3d = np.exp(-(x**2 + y**2) / (2 * sigma_3d**2))
gaussian_3d /= np.sum(gaussian_3d)

# 3D surface plot
fig = plt.figure(figsize=(10, 7))
ax3d = fig.add_subplot(111, projection='3d')
ax3d.plot_surface(x, y, gaussian_3d, cmap='viridis')
ax3d.set_title('51x51 Gaussian Kernel (σ=2)')
ax3d.set_xlabel('X')
ax3d.set_ylabel('Y')
ax3d.set_zlabel('Kernel Coefficient')
plt.show()
```
