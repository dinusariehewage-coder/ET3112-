```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

sigma = 2
size = 5
k = size // 2

# Coordinate grid
x, y = np.meshgrid(np.arange(-k, k+1), np.arange(-k, k+1))

# Gaussian
G = (1 / (2 * np.pi * sigma**2)) * np.exp(-(x**2 + y**2) / (2 * sigma**2))

# Derivatives
Gx = -(x / sigma**2) * G
Gy = -(y / sigma**2) * G

# Normalize kernels (sum of absolute values = 1)
Gx = Gx / np.sum(np.abs(Gx))
Gy = Gy / np.sum(np.abs(Gy))

print("5×5 DoG kernel (x-direction):\n", Gx)
print("\n5×5 DoG kernel (y-direction):\n", Gy)

```
