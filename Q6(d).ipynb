```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

img = cv.imread('/Assignment 1/runway.png', cv.IMREAD_GRAYSCALE)

if img is None:
    raise FileNotFoundError("Check image path.")

# Apply convolution
grad_x = cv.filter2D(img, cv.CV_64F, Gx)
grad_y = cv.filter2D(img, cv.CV_64F, Gy)

# Display
plt.figure(figsize=(12, 4))

plt.subplot(1, 3, 1)
plt.imshow(img, cmap='gray')
plt.title('Original')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(np.abs(grad_x), cmap='gray')
plt.title('Gradient X (DoG)')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(np.abs(grad_y), cmap='gray')
plt.title('Gradient Y (DoG)')
plt.axis('off')

plt.show()
```
