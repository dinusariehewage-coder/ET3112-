```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt


# LOAD IMAGE  (EDIT PATH ONLY)

image_path = '/Assignment 1/emma_salt_pepper.jpg'

img = cv.imread(image_path, cv.IMREAD_GRAYSCALE)

if img is None:
    raise FileNotFoundError("Check image path")


# (a) Gaussian Smoothing

gaussian_filtered = cv.GaussianBlur(img, (5, 5), sigmaX=1)


# (b) Median Filtering

median_filtered = cv.medianBlur(img, 5)


# Display Results

plt.figure(figsize=(12, 4))

plt.subplot(1, 3, 1)
plt.imshow(img, cmap='gray')
plt.title('Noisy Image (Salt & Pepper)')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(gaussian_filtered, cmap='gray')
plt.title('Gaussian Smoothing')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(median_filtered, cmap='gray')
plt.title('Median Filtering')
plt.axis('off')

plt.show()
```
